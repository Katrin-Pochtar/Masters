```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, LSTM, Dense
from tensorflow.keras.utils import plot_model
import matplotlib.pyplot as plt
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
from statsmodels.graphics.tsaplots import plot_acf


def load_data(file_path='data/data_for_modeling.csv'):
    df = pd.read_csv(file_path, parse_dates=['date'])
    df.set_index('date', inplace=True)
    return df[['unique_id', 'units']]


def train_holdout_split(series, season_horizon, month_horizon, week_horizon):
    n = len(series)
    if n < 2:
        raise ValueError(f"Series too short for ID {series.name}: need ≥2 points, got {n}")
    season_h = min(season_horizon, n - 1)
    month_h = min(month_horizon, season_h)
    week_h = min(week_horizon, month_h)
    train = series.iloc[: n - season_h]
    season = series.iloc[-season_h:]
    month = series.iloc[-month_h:]
    week = series.iloc[-week_h:]
    return train, week, month, season


def create_multistep_sequences(data, window, output_steps):
    X, y = [], []
    n = len(data)
    for i in range(n - window - output_steps + 1):
        X.append(data[i:i+window])
        y.append(data[i+window:i+window+output_steps].flatten())
    if not X:
        raise ValueError(f"Need window+output_steps={window+output_steps}, got {n} points")
    return np.stack(X), np.stack(y)


def build_dataset_and_holdouts(df, window, season_horizon, month_horizon, week_horizon):
    X_list, y_list = [], []
    holdouts = {'Week': {}, 'Month': {}, 'Season': {}}
    for uid, group in df.groupby('unique_id'):
        series = group['units'].sort_index()
        train, w, m, s = train_holdout_split(series, season_horizon, month_horizon, week_horizon)
        scaled = train.values.reshape(-1,1)
        X, y = create_multistep_sequences(scaled, window, season_horizon)
        X_list.append(X)
        y_list.append(y)
        holdouts['Week'][uid] = w.values
        holdouts['Month'][uid] = m.values
        holdouts['Season'][uid] = s.values
    X_all = np.concatenate(X_list, axis=0)
    y_all = np.concatenate(y_list, axis=0)
    return X_all, y_all, holdouts


def build_model(window, output_steps, features=1, lstm_units=50):
    model = Sequential([
        Input(shape=(window, features)),
        LSTM(lstm_units, activation='tanh'),
        Dense(output_steps)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model


def evaluate_holdouts(model, df, holdouts, window):
    metrics = {'Week': {}, 'Month': {}, 'Season': {}}
    for horizon, series_dict in holdouts.items():
        for uid, true_vals in series_dict.items():
            series = df[df['unique_id'] == uid]['units'].sort_index()
            train, _, _, _ = train_holdout_split(
                series,
                season_horizon=len(true_vals) if horizon=='Season' else len(true_vals),
                month_horizon=len(true_vals) if horizon=='Month' else len(true_vals),
                week_horizon=len(true_vals) if horizon=='Week' else len(true_vals)
            )
            seq = train.values.reshape(-1,1)[-window:][None,...]
            pred = model.predict(seq).flatten()
            pred_vals = pred[:len(true_vals)] if horizon in ['Week','Month'] else pred
            mse = mean_squared_error(true_vals, pred_vals)
            rmse = np.sqrt(mse)
            mae = mean_absolute_error(true_vals, pred_vals)
            mape = mean_absolute_percentage_error(true_vals, pred_vals)
            r2 = r2_score(true_vals, pred_vals)
            metrics[horizon][uid] = {'MSE':mse,'RMSE':rmse,'MAE':mae,'MAPE':mape,'R2':r2}
    return metrics


def explain_deepshap(model, X_background, X_test):
    import shap
    try:
        explainer = shap.DeepExplainer(model, X_background)
        shap_values = explainer.shap_values(X_test)
        shap.summary_plot(shap_values, X_test, show=False)
        plt.title('SHAP Summary'); plt.tight_layout(); plt.show()
        try:
            shap.initjs()
            _ = shap.force_plot(explainer.expected_value[0], shap_values[0][0], matplotlib=True)
            plt.show()
        except Exception:
            print("Unable to generate force plot in this environment.")
    except LookupError:
        print("DeepExplainer not supported; skipping SHAP explanation.")


def main(
    file_path='data/data_for_modeling.csv',
    window_candidates=[14,30,60,90,120],
    week_horizon=7,
    month_horizon=30,
    season_horizon=90,
    epochs=20,
    batch_size=32,
    validation_split=0.1
):
    df = load_data(file_path)
    print(f"Loaded {df['unique_id'].nunique()} series, {len(df)} rows.")

    first_uid = df['unique_id'].iloc[0]
    first_series = df[df['unique_id']==first_uid]['units'].sort_index()
    tr,_,_,_ = train_holdout_split(first_series,season_horizon,month_horizon,week_horizon)
    plot_acf(tr,lags=120)
    plt.title(f"ACF for {first_uid}"); plt.show()

    best_w, best_loss = None, np.inf
    for w in window_candidates:
        X_tmp, y_tmp, holdouts = build_dataset_and_holdouts(df, w, season_horizon, month_horizon, week_horizon)
        split_idx = int(len(X_tmp)*(1-validation_split))
        X_tr, X_val = X_tmp[:split_idx], X_tmp[split_idx:]
        y_tr, y_val = y_tmp[:split_idx], y_tmp[split_idx:]
        mdl = build_model(w, season_horizon)
        hist = mdl.fit(X_tr, y_tr, epochs=epochs, batch_size=batch_size,
                       validation_data=(X_val,y_val), shuffle=False, verbose=0)
        loss = hist.history['val_loss'][-1]
        print(f"Window={w}: val_loss={loss:.4f}")
        if loss < best_loss:
            best_loss, best_w = loss, w
    print(f"Optimal window: {best_w} (val_loss={best_loss:.4f})")

    X, y, holdouts = build_dataset_and_holdouts(df, best_w, season_horizon, month_horizon, week_horizon)
    print(f"Training on window={best_w}: X={X.shape}, y={y.shape}")
    model = build_model(best_w, season_horizon)
    plot_model(model,'model_architecture.png',show_shapes=True)
    print("Saved model_architecture.png")
    model.summary()
    split_idx = int(len(X)*(1-validation_split))
    X_tr, X_val = X[:split_idx], X[split_idx:]
    y_tr, y_val = y[:split_idx], y[split_idx:]
    model.fit(X_tr, y_tr, epochs=epochs, batch_size=batch_size,
              validation_data=(X_val,y_val), shuffle=False, verbose=2)
    model.save('forecast_model.h5')
    print("Saved model to 'forecast_model.h5'.")

    metrics = evaluate_holdouts(model,df,holdouts,best_w)
    for horizon,series_metrics in metrics.items():
        print(f"\n{horizon} holdout metrics:")
        for uid,m in series_metrics.items():
            print(
                f" {uid}: MSE={m['MSE']:.4f}, RMSE={m['RMSE']:.4f}, MAE={m['MAE']:.4f}, "
                f"MAPE={m['MAPE']:.2%}, R2={m['R2']:.4f}"
            )

    bg = X[:min(100,len(X))]
    last_seq = X[-1:]
    explain_deepshap(model,bg,last_seq)

if __name__=='__main__':
    main()
```

```python
import tensorflow as tf
# Clear previous models/graphs
tf.keras.backend.clear_session()

import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
import tensorflow.keras.backend as K
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
from statsmodels.graphics.tsaplots import plot_acf


def load_data(file_path='data/data_for_modeling.csv'):
    df = pd.read_csv(file_path, parse_dates=['date'])
    df.set_index('date', inplace=True)
    return df[['unique_id', 'units']]


def train_holdout_split(series, season_horizon, month_horizon, week_horizon):
    n = len(series)
    if n < 2:
        raise ValueError(f"Series too short for ID {series.name}: need ≥2 points, got {n}")
    season_h = min(season_horizon, n - 1)
    month_h  = min(month_horizon, season_h)
    week_h   = min(week_horizon, month_h)

    train  = series.iloc[: n - season_h]
    season = series.iloc[-season_h:]
    month  = series.iloc[-month_h:]
    week   = series.iloc[-week_h:]
    return train, week, month, season


def create_multitask_sequences(data, window, w_h, m_h, s_h):
    """
    Build sequences and multi-horizon targets:
      X: (num_samples, window, 1)
      y_week: (num_samples, w_h)
      y_month: (num_samples, m_h)
      y_season: (num_samples, s_h)
    """
    X_list, y_w, y_m, y_s = [], [], [], []
    n = len(data)
    for i in range(n - window - s_h + 1):
        seq = data[i:i+window].flatten()
        fut = data[i+window:i+window+s_h].flatten()
        X_list.append(seq)
        y_w.append(fut[:w_h])
        y_m.append(fut[:m_h])
        y_s.append(fut)

    X  = np.stack(X_list)[..., None]
    y_w = np.stack(y_w)
    y_m = np.stack(y_m)
    y_s = np.stack(y_s)
    return X, y_w, y_m, y_s


def build_dataset_and_holdouts(df, window, season_horizon, month_horizon, week_horizon):
    X_list, yw_list, ym_list, ys_list = [], [], [], []
    holdouts = {'Week': {}, 'Month': {}, 'Season': {}}

    for uid, group in df.groupby('unique_id'):
        series = group['units'].sort_index()
        train, w, m, s = train_holdout_split(
            series, season_horizon, month_horizon, week_horizon
        )
        values = train.values
        X, yw, ym, ys = create_multitask_sequences(
            values, window, week_horizon, month_horizon, season_horizon
        )
        X_list.append(X)
        yw_list.append(yw)
        ym_list.append(ym)
        ys_list.append(ys)
        holdouts['Week'][uid]   = w.values
        holdouts['Month'][uid]  = m.values
        holdouts['Season'][uid] = s.values

    X_all   = np.concatenate(X_list, axis=0)
    y_w_all = np.concatenate(yw_list, axis=0)
    y_m_all = np.concatenate(ym_list, axis=0)
    y_s_all = np.concatenate(ys_list, axis=0)
    return X_all, y_w_all, y_m_all, y_s_all, holdouts


def r2_keras(y_true, y_pred):
    """Coefficient of determination metric."""
    SS_res = K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return 1 - SS_res / (SS_tot + K.epsilon())


def build_multihead_model(window, week_horizon, month_horizon, season_horizon,
                          lstm_units=50):
    """Multi-task LSTM with heads for week, month, season."""
    inp = Input(shape=(window, 1), name='input')
    x = LSTM(lstm_units, activation='tanh', name='lstm')(inp)

    out_week   = Dense(week_horizon,  name='week')(x)
    out_month  = Dense(month_horizon, name='month')(x)
    out_season = Dense(season_horizon,name='season')(x)

    model = Model(inputs=inp,
                  outputs=[out_week, out_month, out_season],
                  name='multitask_lstm')
    model.compile(
        optimizer='adam',
        loss={
            'week':   'mse',
            'month':  'mse',
            'season': 'mse'
        },
        loss_weights={
            'week':   1.0,
            'month':  1.0,
            'season': 1.0
        },
        metrics={
            'week':   [
                tf.keras.metrics.RootMeanSquaredError(name='rmse'),
                tf.keras.metrics.MeanAbsoluteError(name='mae'),
                r2_keras
            ],
            'month':  [
                tf.keras.metrics.RootMeanSquaredError(name='rmse'),
                tf.keras.metrics.MeanAbsoluteError(name='mae'),
                r2_keras
            ],
            'season': [
                tf.keras.metrics.RootMeanSquaredError(name='rmse'),
                tf.keras.metrics.MeanAbsoluteError(name='mae'),
                r2_keras
            ],
        }
    )
    return model


def evaluate_holdouts_multi(model, df, holdouts, window):
    metrics = {'Week': {}, 'Month': {}, 'Season': {}}
    for horizon, series_dict in holdouts.items():
        for uid, true_vals in series_dict.items():
            series = df[df['unique_id']==uid]['units'].sort_index()
            train, w, m, s = train_holdout_split(
                series,
                season_horizon=len(true_vals) if horizon=='Season' else len(true_vals),
                month_horizon=len(true_vals) if horizon=='Month' else len(true_vals),
                week_horizon=len(true_vals) if horizon=='Week' else len(true_vals)
            )
            seq = train.values[-window:].reshape(1,window,1)
            pred_w, pred_m, pred_s = model.predict(seq)
            preds = {'Week': pred_w.flatten(), 'Month': pred_m.flatten(), 'Season': pred_s.flatten()}[horizon]

            mse  = mean_squared_error(true_vals, preds)
            rmse = np.sqrt(mse)
            mae  = mean_absolute_error(true_vals, preds)
            mape = mean_absolute_percentage_error(true_vals, preds)
            r2   = r2_score(true_vals, preds)
            metrics[horizon][uid] = {'MSE':mse, 'RMSE':rmse,
                                     'MAE':mae, 'MAPE':mape, 'R2':r2}
    return metrics

# Example usage
if __name__=='__main__':
    df = load_data('data/data_for_modeling.csv')
    window = 30
    week_horizon = 7
    month_horizon = 30
    season_horizon = 90

    X, y_w, y_m, y_s, holdouts = build_dataset_and_holdouts(
        df, window, season_horizon, month_horizon, week_horizon
    )

    model = build_multihead_model(window, week_horizon, month_horizon, season_horizon)
    model.summary()

    model.fit(
        X,
        {'week':   y_w,
         'month':  y_m,
         'season': y_s},
        epochs=20, batch_size=32, validation_split=0.1
    )

    metrics = evaluate_holdouts_multi(model, df, holdouts, window)
    print(metrics)
```

```python
import tensorflow as tf
# Clear previous models/graphs
tf.keras.backend.clear_session()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense
import tensorflow.keras.backend as K
from sklearn.metrics import (
    mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
from statsmodels.graphics.tsaplots import plot_acf


def load_data(file_path='data/data_for_modeling.csv'):
    df = pd.read_csv(file_path, parse_dates=['date'])
    df.set_index('date', inplace=True)
    return df[['unique_id', 'units']]


def train_holdout_split(series, season_horizon, month_horizon, week_horizon):
    n = len(series)
    if n < 2:
        raise ValueError(f"Series too short for ID {series.name}: need ≥2 points, got {n}")
    season_h = min(season_horizon, n - 1)
    month_h  = min(month_horizon, season_h)
    week_h   = min(week_horizon, month_h)

    train  = series.iloc[: n - season_h]
    season = series.iloc[-season_h:]
    month  = series.iloc[-month_h:]
    week   = series.iloc[-week_h:]
    return train, week, month, season


def create_multitask_sequences(data, window, w_h, m_h, s_h):
    X_list, y_w, y_m, y_s = [], [], [], []
    n = len(data)
    for i in range(n - window - s_h + 1):
        seq = data[i:i+window].flatten()
        fut = data[i+window:i+window+s_h].flatten()
        X_list.append(seq)
        y_w.append(fut[:w_h])
        y_m.append(fut[:m_h])
        y_s.append(fut)
    X  = np.stack(X_list)[..., None]
    y_w = np.stack(y_w)
    y_m = np.stack(y_m)
    y_s = np.stack(y_s)
    return X, y_w, y_m, y_s


def build_dataset_and_holdouts(df, window, season_horizon, month_horizon, week_horizon):
    X_list, yw_list, ym_list, ys_list = [], [], [], []
    holdouts = {'Week': {}, 'Month': {}, 'Season': {}}
    for uid, group in df.groupby('unique_id'):
        series = group['units'].sort_index()
        train, w, m, s = train_holdout_split(series, season_horizon, month_horizon, week_horizon)
        X, yw, ym, ys = create_multitask_sequences(
            train.values, window, week_horizon, month_horizon, season_horizon
        )
        X_list.append(X); yw_list.append(yw); ym_list.append(ym); ys_list.append(ys)
        holdouts['Week'][uid]=w.values
        holdouts['Month'][uid]=m.values
        holdouts['Season'][uid]=s.values
    X_all = np.concatenate(X_list,0)
    y_w_all = np.concatenate(yw_list,0)
    y_m_all = np.concatenate(ym_list,0)
    y_s_all = np.concatenate(ys_list,0)
    return X_all, y_w_all, y_m_all, y_s_all, holdouts


def r2_keras(y_true, y_pred):
    # cast to float to avoid dtype mismatch
    y_true = K.cast(y_true, K.floatx())
    y_pred = K.cast(y_pred, K.floatx())
    SS_res = K.sum(K.square(y_true - y_pred))
    SS_tot = K.sum(K.square(y_true - K.mean(y_true)))
    return 1 - SS_res / (SS_tot + K.epsilon())


def build_multihead_model(window, week_horizon, month_horizon, season_horizon, lstm_units=50):
    inp = Input(shape=(window,1), name='input')
    x = LSTM(lstm_units, activation='tanh', name='lstm')(inp)
    out_w = Dense(week_horizon, name='week')(x)
    out_m = Dense(month_horizon, name='month')(x)
    out_s = Dense(season_horizon, name='season')(x)
    model = Model(inp, [out_w, out_m, out_s], name='multitask_lstm')
    model.compile(
        optimizer='adam',
        loss={'week':'mse','month':'mse','season':'mse'},
        loss_weights={'week':1.0,'month':1.0,'season':1.0},
        metrics={
            'week':  [tf.keras.metrics.MeanSquaredError(name='mse')],
            'month': [tf.keras.metrics.MeanSquaredError(name='mse')],
            'season':[tf.keras.metrics.MeanSquaredError(name='mse')]
        }
    )
    return model


def evaluate_holdouts_multi(model, df, holdouts, window):
    metrics = {'Week':{},'Month':{},'Season':{}}
    for horizon, series_dict in holdouts.items():
        for uid, true_vals in series_dict.items():
            series = df[df['unique_id']==uid]['units'].sort_index()
            train, w, m, s = train_holdout_split(
                series,
                season_horizon=len(true_vals) if horizon=='Season' else len(true_vals),
                month_horizon=len(true_vals) if horizon=='Month' else len(true_vals),
                week_horizon=len(true_vals) if horizon=='Week' else len(true_vals)
            )
            seq = train.values[-window:].reshape(1,window,1)
            pw, pm, ps = model.predict(seq)
            preds = {'Week':pw.flatten(),'Month':pm.flatten(),'Season':ps.flatten()}[horizon]
            mse=mean_squared_error(true_vals,preds)
            rmse=np.sqrt(mse)
            mae=mean_absolute_error(true_vals,preds)
            mape=mean_absolute_percentage_error(true_vals,preds)
            r2=r2_score(true_vals,preds)
            metrics[horizon][uid]={'MSE':mse,'RMSE':rmse,'MAE':mae,'MAPE':mape,'R2':r2}
    return metrics

# Example usage
if __name__=='__main__':
    df = load_data()
    window, w_h, m_h, s_h = 90, 7, 30, 90
    X, y_w, y_m, y_s, holdouts = build_dataset_and_holdouts(df, window, s_h, m_h, w_h)

    model = build_multihead_model(window, w_h, m_h, s_h)
    history = model.fit(
        X, {'week':y_w,'month':y_m,'season':y_s},
        epochs=20, batch_size=32, validation_split=0.1
    )

    # Plot training curves
    plt.figure(figsize=(12,4))
    plt.subplot(1,2,1)
    plt.plot(history.history['loss'], label='loss')
    plt.plot(history.history['val_loss'], label='val_loss')
    plt.title('Loss'); plt.legend()
    plt.subplot(1,2,2)
    plt.plot(history.history['season_rmse'], label='season_rmse')
    plt.plot(history.history['val_season_rmse'], label='val_season_rmse')
    plt.title('Season RMSE'); plt.legend()
    plt.show()

    metrics = evaluate_holdouts_multi(model, df, holdouts, window)
    for horizon, m in metrics.items():
        df_met = pd.DataFrame.from_dict(m, orient='index')
        print(f"\n{horizon} Holdout Metrics")
        print(df_met)
```
