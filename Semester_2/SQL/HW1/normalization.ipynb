Исходная таблица транзакций содержала избыточную информацию о продукте (бренд, цена, и прочее), что нарушало принципы первой нормальной формы (1НФ), а также приводило к частичным зависимостям (2НФ) и транзитивным зависимостям (3НФ). Для устранения этих проблем было принято решение выделить информацию о продуктах в отдельную таблицу Product.

Поскольку поле product_id оказалось неуникальным, был введён ключ custom_product_id, собранный из:

custom_product_id = 'id_' + product_id + '_list_price_' + list_price


Это позволило гарантировать уникальность записей в таблице product, так как каждая комбинация идентификатора продукта и цены становится уникальной.

Кроме того, при анализе данных в таблице транзакций выяснилось, что запись с customer_id 5034 не имеет соответствия в таблице customer. В связи с этим данная запись была удалена для обеспечения целостности данных.

В результате разделения данных на три таблицы — customer, product и transaction — каждая таблица теперь содержит атрибуты, зависящие исключительно от своего первичного ключа. Это позволяет привести базу данных к третьей нормальной форме (3НФ), устраняя избыточность и обеспечивая корректное обновление данных.
