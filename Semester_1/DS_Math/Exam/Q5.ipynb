```python
import numpy as np
import matplotlib.pyplot as plt
```

# Linear regression

```python
x = np.array([1, 2, 3, 4, 5], dtype=float)
y = np.array([2, 4, 6, 8, 10], dtype=float)
```

```python
learning_rate = 0.01
num_epochs = 1000

w = 0.0
b = 0.0

n = len(x)

def mse(y_true, y_pred):
    return np.mean((y_true - y_pred)**2)
```

```python
loss_history = []

for epoch in range(num_epochs):
    # 1) Предсказание модели
    y_pred = w * x + b  # вектор предсказаний

    # 2) Вычисление ошибки
    error = y_pred - y  # вектор ошибок (y_pred - y)
    loss = mse(y, y_pred)
    
    # Сохраняем в историю (можно потом визуализировать)
    loss_history.append(loss)

    # 3) Вычисление градиентов
    # dMSE/dw = (2/n) * sum( (w*x_i + b - y_i) * x_i )
    dw = (2/n) * np.sum(error * x)
    # dMSE/db = (2/n) * sum( (w*x_i + b - y_i) )
    db = (2/n) * np.sum(error)

    # 4) Обновление параметров (шаг градиентного спуска)
    w -= learning_rate * dw
    b -= learning_rate * db
    
print(f"Final w = {w:.4f}")
print(f"Final b = {b:.4f}")
print(f"Final MSE = {loss_history[-1]:.6f}")

y_pred_final = w * x + b
```

# Plot

```python
plt.figure(figsize=(6, 4))

plt.scatter(x, y, color='blue', label='Исходные точки')
plt.plot(x, y_pred_final, color='red', label='Линия регрессии')

plt.title("Линейная регрессия с помощью градиентного спуска")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(range(num_epochs), loss_history, color='green')
plt.title("MSE во время обучения (Gradient Descent)")
plt.xlabel("Номер эпохи")
plt.ylabel("MSE")
plt.grid(True)
plt.show()
```
