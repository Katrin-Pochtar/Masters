```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sigmoid function

```python
def sigma(x):
    return 1 / (1 + np.exp(-x))
```

```python
x_values = np.linspace(-10, 10, 200)
sig_values = sigma(x_values)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x_values, sig_values, label=r'$\sigma(x)$')
plt.title("Логистическая функция (сигмоида)")
plt.xlabel("x")
plt.ylabel(r'$\sigma(x)$')
plt.grid(True)
plt.legend()
plt.show()
```

# Binary Cross Entropy

```python
p_values = sigma(x_values)
loss_y1 = -np.log(p_values)
loss_y0 = -np.log(1 - p_values)
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x_values, loss_y1, label=r'BCE$(y=1)$ = $-\ln(\sigma(x))$')
plt.plot(x_values, loss_y0, label=r'BCE$(y=0)$ = $-\ln(1-\sigma(x))$')

plt.title("Бинарная кросс-энтропия для y=0 и y=1")
plt.xlabel("x")
plt.ylabel("Loss")
plt.ylim(-0.1, 5)
plt.legend()
plt.show()
```
