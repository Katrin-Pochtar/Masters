```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
X = np.random.normal(5, 2, 100)
Y = 2 * X + np.random.normal(0, 1, 100)
```

# Manual

```python
n = len(X)

mean_X = np.mean(X)
mean_Y = np.mean(Y)

cov_manual = sum((X[i] - mean_X)*(Y[i] - mean_Y) for i in range(n)) / (n - 1)

var_X = sum((X[i] - mean_X)**2 for i in range(n)) / (n - 1)
var_Y = sum((Y[i] - mean_Y)**2 for i in range(n)) / (n - 1)

corr_manual = cov_manual / (var_X**0.5 * var_Y**0.5)

print("Ковариация =", cov_manual)
print("Корреляция =", corr_manual)
```

# Numpy

```python
cov_numpy = np.cov(X, Y, ddof=1)[0, 1]
corr_numpy = np.corrcoef(X, Y)[0, 1]

print("Ковариация (np.cov) =", cov_numpy)
print("Корреляция (np.corrcoef) =", corr_numpy)
```

# Scatter plot

```python
sns.scatterplot(x=X, y=Y)

plt.title("Зависимость X и Y (Seaborn Scatter)")
plt.grid(True)
plt.show()
```
